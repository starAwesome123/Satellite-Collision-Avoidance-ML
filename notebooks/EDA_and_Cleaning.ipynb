```python
df1 = pd.read_csv('train_data.csv')
print(df1.shape)
o=df1['c_object_type']
o.shape
```

```python
filtered_df = df1[df1['c_object_type'].isin(['DEBRIS', 'UNKNOWN', 'PAYLOAD', 'ROCKET BODY', 'TBA'])]

# Count the occurrences of each value
value_counts = filtered_df['c_object_type'].value_counts()

# Print the counts
print(value_counts)
```

```python
value_counts_df = value_counts.reset_index(name='count')

total_count = value_counts_df['count'].sum()

# Calculate percentages
value_counts_df['percentage'] = (value_counts_df['count'] / total_count) * 100

# Create the bar plot
sns.barplot(x='c_object_type', y='count', data=value_counts_df)

# Set the title, x-label, and y-label
plt.title('Distribution of Chaser Object Types')
plt.xlabel('Chaser Object Type')
plt.ylabel('Count')

# Annotate percentages
for i in range(len(value_counts_df)):
    plt.text(i, value_counts_df['count'][i], f"{value_counts_df['percentage'][i]:.2f}%", ha="center")

# Show the plot
plt.show()
```

```python
#rehab
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
train_data= pd.read_csv('train_data.csv')
train_data.head()
#Select Frist 14 features

df_copy = train_data[['event_id', 'time_to_tca', 'mission_id', 'risk', 'max_risk_estimate', 'max_risk_scaling', 'miss_distance', 'relative_speed', 'relative_position_r', 'relative_position_t', 'relative_position_n', 'relative_velocity_r', 'relative_velocity_t', 'relative_velocity_n','c_j2k_inc']]
df_copy
cdm_per_event = df_copy.groupby('event_id').size().reset_index(name='cdm_count')
cdm_count_distribution = cdm_per_event['cdm_count'].value_counts().sort_index()
median_cdm_count = cdm_per_event['cdm_count'].median()
print(f"Median of CDM Count per Event: {median_cdm_count}")

# Step 3: Plot the distribution of CDM counts per event
plt.figure(figsize=(6, 6))
sns.barplot(x=cdm_count_distribution.index, y=cdm_count_distribution.values, color='blue')
plt.axvline(median_cdm_count, color='red', linestyle='--', label=f'Median: {median_cdm_count}')
plt.legend()
# Step 4: Set the labels and title
plt.xlabel('Number of CDMs per Event')
plt.ylabel('Count of Events')
plt.title('Count of CDMs per Event')

# Show the plot
plt.show()
```
